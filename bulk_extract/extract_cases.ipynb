# Bulk case download and extract

```python
import os
import sys
sys.path.append('..')

import lzma
import json
import pandas as pd

from config import settings
import utils
```

```python
jurisdiction_to_load = 'Illinois'
dir_exists = False
dir_path = None
for filename in os.listdir(settings.DATA_DIR):
    if jurisdiction_to_load in filename:
        if os.path.exists(os.path.join(settings.DATA_DIR, filename + '/data/data.jsonl.xz')):
            dir_exists = True
            break
if dir_exists:
    dir_path = os.path.join(settings.DATA_DIR, filename)
else:
    dir_path = utils.get_cases_from_bulk('Illinois')
```

```python
compressed_file = os.path.join(settings.DATA_DIR, dir_path + '/data/data.jsonl.xz')
os.path.exists(compressed_file)
```

```python
cases = []
print(compressed_file)
with lzma.open(compressed_file) as infile:
    for line in infile:
        record = json.loads(str(line, 'utf-8'))
        cases.append(record)

print("Case count: %s" % len(cases))
```

```python
df = pd.DataFrame(cases)
df.head()
```
